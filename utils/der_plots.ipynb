```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x,y = np.meshgrid(np.linspace(0,1,100), np.linspace(0,1,100))
```

```python
#x,y =np.linspace(0,1,100),np.linspace(0,1,100)
```

```python
z = (10 * np.sin(y*2) + 10 * x - 12.0) * (x * y * (x-1) * (y-1)) * 15
```

```python
dydx = 30*(y-1)*y*(x*(15*x+10*np.sin(2*y)-22)-5*np.sin(2*y)+6)
```

```python
dydy = 30*(x-1)*x*(5*(2*y-1)*np.sin(2*y)+2*y*(5*((y-1)*np.cos(2*y)+x)-6)-5*x+6)
```

```python
from matplotlib.gridspec import GridSpec
```

```python
subplot_size = 6.
gs = GridSpec(1, 4)
(width, height) = (4*subplot_size, 1*subplot_size) #5 col and 3 rows, each of size subplot_size
```

```python
fig = plt.figure(figsize = (width, height))
plt.subplot(gs[0, 0])
plt.title('$z$')
plt.contourf(x,y,z)
plt.subplot(gs[0, 1])
plt.title('$dz/dx$')
plt.contourf(x,y,dydx)
plt.subplot(gs[0, 2])
plt.title('$dz/dy$')
plt.contourf(x,y,dydy)
plt.subplot(gs[0, 3])
plt.title('$sqrt((dz/dx)^2 + (dz/dy)^2$')
plt.contourf(x,y,np.sqrt(dydx**2 + dydy**2))
```

```python
gs = GridSpec(1,4)
plt.figure(figsize=(4 * 3, 1 * 3))
plt.subplot(gs[0,0])
plt.contourf(x,y,z)
plt.subplot(gs[0,1])
plt.contourf(x,y,dydx)
```
